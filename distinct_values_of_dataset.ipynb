```python
# This code is used to get the distinct values of a column in a CSV file.
# It reads the CSV file, gets the distinct values for each column, and outputs them as a list with the new column name in each line.
# The output is a list of strings, where each string contains the column name and its distinct values.
```

```python
import pandas as pd

def get_distinct_values(csv_file_path):
    # Read the CSV file
    df = pd.read_csv(csv_file_path)
    
    # Get distinct values for each column and output as a list with new column name in each line
    distinct_values = {col: df[col].unique() for col in df.columns}
    
    output = []
    for col, values in distinct_values.items():
        output.append(f"{col}: {list(values)}")
    
    return output
```

```python
csv_file_path = 'sample_data/working_database/final_income_sheet_new_seq.csv'
get_distinct_values(csv_file_path)
```
